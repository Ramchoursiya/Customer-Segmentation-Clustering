```python
plt.figure(figsize = (10,50))
for i in range(len(data.columns)):
    plt.subplot(17,1,i + 1)
    sns.distplot(data[data.columns[i]],kde_kws = {"color":"b","lw":3,"label":"KDE"},hist_kws = {'color':"g"})
    plt.title(data.columns[i])
plt.tight_layout()
```

```python
from sklearn.preprocessing import StandardScaler,normalize
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
scaler = StandardScaler()
```

```python
data_scaled = scaler.fit_transform(data)
```

```python
data_scaled
```

```python
kmeans = KMeans(7)
kmeans.fit(data_scaled)
labels = kmeans.labels_
```

```python
kmeans.cluster_centers_.shape
```

```python
cluster_centers = pd.DataFrame(data = kmeans.cluster_centers_,columns = [data.columns])
cluster_centers
```

```python
cluster_centers = scaler.inverse_transform(cluster_centers)
cluster_centers = pd.DataFrame(data = cluster_centers,columns = [data.columns])
cluster_centers
```

```python
labels.shape
```

```python
labels.max()
```

```python
y_kmeans = kmeans.fit_predict(data_scaled)
y_kmeans
```

```python
data_clustered = pd.concat([data,pd.DataFrame({'cluster':labels})],axis = 1)
data_clustered.head()
```

```python
for i in data.columns:
    plt.figure(figsize= (35,5))
    for j in range(7):
        plt.subplot(1,7,j + 1)
        cluster = data_clustered[data_clustered['cluster'] == j]
        cluster[i].hist(bins  = 20)
        plt.title('{} \nCluster {}'.format(i,j))
    plt.show()
```

```python
pca = PCA(n_components=2)
principal_comp = pca.fit_transform(data_scaled)
principal_comp
```

```python
pca_df = pd.DataFrame(data = principal_comp,columns = ['pca1','pca2'])
pca_df.head()
```

```python
pca_df = pd.concat([pca_df,pd.DataFrame({'cluster':labels})],axis = 1)
pca_df.head()
```

```python
plt.figure(figsize = (10,10))
ax = sns.scatterplot(x ="pca1",y = "pca2",hue = 'cluster',data = pca_df,palette=['red','green','blue','pink','yellow','gray','purple'])
plt.show()
```
